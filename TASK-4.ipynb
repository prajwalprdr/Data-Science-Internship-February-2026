```python
# Function to validate recharge amount
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    
    if amount < 50:
        print("âŒ Recharge amount must be at least â‚¹50.")
        return False
    elif amount not in valid_plans:
        print("âŒ Invalid plan selected.")
        print("Available plans are:", valid_plans)
        return False
    else:
        print("âœ… Recharge successful for â‚¹", amount)
        return True


# Main program with retry option
while True:
    try:
        recharge_amount = int(input("Enter recharge amount: â‚¹"))
        
        if validate_recharge(recharge_amount):
            break   # Exit loop if recharge is valid
        
        print("Please try again.\n")
    
    except ValueError:
        print("âŒ Please enter a valid numeric amount.\n")
```

```python
# Function to check stock levels
def check_inventory(products):
    for product, stock in products.items():
        if stock < 15:
            print(f"{product} â†’ Reorder Alert (Stock: {stock})")
        else:
            print(f"{product} â†’ Stock OK (Stock: {stock})")


# Sample inventory dictionary
inventory = {
    "Rice": 25,
    "Sugar": 10,
    "Oil": 18,
    "Salt": 8,
    "Wheat": 30
}

# Call function
check_inventory(inventory)
```

```python
# Function to process student result
def process_result(marks):
    total = 0
    
    # Calculate total using loop
    for mark in marks:
        total += mark
    
    average = total / len(marks)
    
    # Determine result
    if average >= 50:
        result = "Pass"
    else:
        result = "Fail"
    
    return average, result


# Sample marks list
student_marks = [60, 45, 70, 55, 40]

# Function call
avg, final_result = process_result(student_marks)

print("Average Marks:", avg)
print("Result:", final_result)
```

```python
# Function to calculate cab fare
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12
    
    fare = base_fare + (distance * per_km_rate)
    
    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra during peak hours
    
    return fare


# Main program with retry option
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")
        
        total_fare = calculate_fare(distance, peak_hour)
        
        print(f"Total Cab Fare: â‚¹{total_fare}")
        
        retry = input("Do you want to calculate again? (yes/no): ")
        if retry.lower() != "yes":
            print("Thank you for using Cab Fare Estimator ðŸš•")
            break
    
    except ValueError:
        print("âŒ Please enter a valid distance.\n")
```

```python
# Function to check attendance eligibility
def check_attendance(attendance_list):
    present_days = 0
    
    # Count present days using loop
    for day in attendance_list:
        if day.upper() == "P":
            present_days += 1
    
    total_days = len(attendance_list)
    percentage = (present_days / total_days) * 100
    
    # Determine eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"
    
    return percentage, status


# Sample attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Function call
percent, result = check_attendance(attendance)

print("Attendance Percentage:", percent, "%")
print("Status:", result)
```

```python
# Function to check password strength
def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Check length
    if len(password) < 8:
        return "Weak Password âŒ (Minimum 8 characters required)"
    
    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
    
    # Check conditions
    if has_digit and has_special:
        return "Strong Password âœ…"
    else:
        return "Weak Password âŒ (Must contain at least one digit and one special character @#$)"


# Example usage
user_password = input("Enter password: ")
result = check_password(user_password)
print(result)
```
