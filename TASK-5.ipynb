```python
# Function to analyze post engagement
def analyze_engagement(likes_list):
    total_likes = 0
    
    # Loop to calculate total likes
    for like in likes_list:
        total_likes += like
    
    print("Total Likes:", total_likes)
    
    # Condition to check post status
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes = [200, 300, 150, 400]   # Total = 1050
analyze_engagement(likes)
```

```python
# Function to check rainfall adequacy
def rainfall_checker(rainfall_list, required_level):
    total_rainfall = 0
    
    # Loop to calculate total rainfall
    for rain in rainfall_list:
        total_rainfall += rain
    
    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_list)
    
    print("Average Rainfall:", int(average_rainfall))
    
    # Check rainfall status
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
rainfall_data = [70, 75, 68, 80, 67]   # Average = 72
required = 70

rainfall_checker(rainfall_data, required)
```

```python
# Function to detect duplicate usernames
def detect_duplicate_accounts(usernames):
    
    unique_usernames = set(usernames)
    
    # If length differs, duplicates exist
    if len(unique_usernames) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
users = ["user1", "user2", "user3", "user1"]
detect_duplicate_accounts(users)
```

```python
# Function to filter premium crop prices
def filter_premium_crops(prices, threshold=2000):
    
    premium_crops = []
    
    # List filtering using loop
    for price in prices:
        if price > threshold:
            premium_crops.append(price)
    
    print("Premium Crops:", premium_crops)


# Example usage
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Function to check application health
def check_application_health(error_count):
    
    print("Error Count:", error_count)
    
    # Conditional checks
    if error_count == 0:
        print("System Status: Healthy")
    elif error_count <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


# Example usage
check_application_health(7)
```

```python
# Function to check daily transaction limit
def check_transaction(amount):
    
    daily_limit = 50000
    print("Transaction Amount:", amount)
    
    # Conditional logic
    if amount <= daily_limit:
        print("Transaction Status: Approved")
    else:
        print("Transaction Status: Rejected")


# Example usage
check_transaction(60000)
```

```python
# Function to calculate attendance percentage and check eligibility
def check_attendance(attendance_list):
    
    total_classes = len(attendance_list)
    attended_classes = 0
    
    # Loop to count attended classes (1 = Present, 0 = Absent)
    for status in attendance_list:
        attended_classes += status
    
    # Calculate percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    
    print("Attendance Percentage:", attendance_percentage)
    
    # Check eligibility
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example usage
attendance_data = [1, 1, 1, 1, 0]  # 4 out of 5 present → 80%
check_attendance(attendance_data)
```

```python
# Function to calculate electricity bill and usage classification
def calculate_electricity_bill(units):
    
    bill = 0
    
    # Slab calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    
    # Usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
    
    return bill, usage_status


# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Units Consumed:", units_consumed)
print("Total Bill: ₹", total_bill)
print("Usage Classification:", status)
```
