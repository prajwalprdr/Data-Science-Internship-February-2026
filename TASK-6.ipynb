```python
def smart_parking_system(capacity, vehicle_logs):
    current_vehicles = 0
    peak_usage = 0

    for log in vehicle_logs:
        if log == "IN":
            current_vehicles += 1
        elif log == "OUT":
            if current_vehicles > 0:
                current_vehicles -= 1
        
        # Track peak usage
        if current_vehicles > peak_usage:
            peak_usage = current_vehicles

    print("Currently Parked Vehicles:", current_vehicles)
    
    if current_vehicles > capacity:
        print("Parking Status: Exceeded Capacity! Alert!")
    else:
        print("Parking Status: Available")
    
    print("Peak Parking Usage:", peak_usage)


# Sample Input
capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

smart_parking_system(capacity, vehicle_logs)
```

```python
def estimate_delivery_time(distance, traffic, weather):
    # Base speed (km per hour)
    speed = 30
    
    # Base time calculation
    base_time = (distance / speed) * 60   # convert to minutes
    total_time = base_time

    # Traffic adjustment
    if traffic == "Low":
        total_time += 5
    elif traffic == "Medium":
        total_time += 10
    elif traffic == "High":
        total_time += 20

    # Weather adjustment
    if weather == "Cloudy":
        total_time += 5
    elif weather == "Rainy":
        total_time += 15

    print("Estimated Delivery Time:", round(total_time), "minutes")


# Sample Input
distance = 8
traffic = "High"
weather = "Rainy"

estimate_delivery_time(distance, traffic, weather)
```

```python
def seat_occupancy_analyzer(total_seats, booked_seats):
    booked_count = len(booked_seats)
    
    # Calculate occupancy percentage
    occupancy = (booked_count / total_seats) * 100
    
    print("Occupancy:", round(occupancy), "%")
    
    # Determine show status
    if occupancy == 100:
        print("Show Status: Housefull")
        print("Suggestion: Open Additional Show")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
        print("Suggestion: Consider Opening Additional Show")
    else:
        print("Show Status: Seats Available")
        print("Suggestion: No Need for Additional Show")


# Sample Input
total_seats = 200
booked_seats = [1] * 150   # 150 seats booked

seat_occupancy_analyzer(total_seats, booked_seats)
```

```python
def classify_server_load(cpu_readings):
    total = 0
    
    # Calculate total using loop
    for reading in cpu_readings:
        total += reading
    
    # Calculate average
    average = total / len(cpu_readings)
    
    print("Average CPU Load:", round(average), "%")
    
    # Classification
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average < 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```python
def classroom_resource_monitor(resource_usage):
    overused = []

    # Check each resource
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)

    # Display results
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

classroom_resource_monitor(resource_usage)
```

```python
def event_registration_controller(capacity, total_registrations):
    confirmed = 0
    waitlisted = 0

    # Process each registration
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlisted += 1

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")


# Sample Input
capacity = 100
registrations = 105

event_registration_controller(capacity, registrations)
```
